# Web Scraping


- Extracting data from websites

## Example
- Let's consider [https://en.wikipedia.org/wiki/Wikipedia:Fundraising_statistics](https://en.wikipedia.org/wiki/Wikipedia:Fundraising_statistics)


```python
import pandas as pd
```

```python
url= "https://en.wikipedia.org/wiki/Wikipedia:Fundraising_statistics"
```

```python
data=pd.read_html(url) 
```

```python
data
```

```python
type(data)
```

```python
len(data)
```

```python
data[0]
```

```python
fundraising = data[0]
```

```python
fundraising.dtypes
```

```python
fundraising["Exp"]= fundraising["Expenses"].str[2:]
```

```python
fundraising["Exp"]
```

```python
fundraising["Exp"]= fundraising["Exp"].str.replace(',','')
```

```python
fundraising["Exp"]
```

```python
fundraising["Exp"]=pd.to_numeric(fundraising["Exp"])
```

```python
fundraising["Exp"]
```

```python
fundraising
```

```python
fundraising.dtypes
```

```python
fundraising["Rev"]= fundraising["Revenue"].str[2:]
fundraising["Rev"]= fundraising["Rev"].str.replace(',','')
fundraising["Rev"]=pd.to_numeric(fundraising["Rev"])
```

```python
fundraising
```

```python
fundraising.dtypes
```

```python
fundraising["AssetRise"]= fundraising["Asset rise"].str[2:]
fundraising["AssetRise"]= fundraising["AssetRise"].str.replace(',','')
fundraising["AssetRise"]=pd.to_numeric(fundraising["AssetRise"])
```

```python
fundraising["TotalAssets"]= fundraising["Total assets"].str[2:]
fundraising["TotalAssets"]= fundraising["TotalAssets"].str.replace(',','')
fundraising["TotalAssets"]=pd.to_numeric(fundraising["TotalAssets"])
```

```python
fundraising
```

```python
del (fundraising["Revenue"])
```

```python
del (fundraising["Total assets"])
```

```python
del (fundraising["Expenses"],fundraising['Asset rise'])
```

```python
fundraising
```

```python
fundraising.isna().any()
```

```python
fundraising.dtypes
```

```python
fundraising["Year"]= fundraising["Year"].str[-4:]
```

```python
fundraising
```

```python
fundraising.set_index('Year',inplace=True)
```

```python
fundraising
```
